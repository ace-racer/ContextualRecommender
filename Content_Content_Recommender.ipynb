```python
import pandas as pd
import numpy as np
from sklearn.neighbors import NearestNeighbors
from sklearn import preprocessing
```

```python
# configurations
num_nghbrs = 5
stream_tag_frequency_location = "data/tag_frequency_streams.csv"
knn_columns_to_select = ["StreamID"]
min_distance_threshold = 0.0003

# maximum distance within which if the distance lies we consider that the pair can be recommended
max_distance_threshold = 4

nearest_neighbors_streams_output_file = "data/nearest_neighbor_streams.csv"

# end of configurations 

# constants
NEWLINE = "\n"
# end of constants

stream_tag_frequency_df = pd.read_csv(stream_tag_frequency_location)
nbrs = NearestNeighbors(n_neighbors = num_nghbrs, algorithm='ball_tree').fit(stream_tag_frequency_df)

# set the testing data source - for now keeping the same

nearest_neighbors_testing_df = stream_tag_frequency_df

# end of setting testing data source

distances, indices = nbrs.kneighbors(nearest_neighbors_testing_df)

def getTextForOutput(stream_details):
        text = ""
        if stream_details is not None:
            for column in knn_columns_to_select:
                text += str(stream_details[column]) + ","
        return text

def getColumnNames():
        """Get the column names for the output file"""
        column_names = ""

        # Generate column names for the target from the test data
        for column in knn_columns_to_select:
            column_names += "Target " + column + ","


        for itr in range(0, num_nghbrs):
            for column in knn_columns_to_select:
                column_names += "Neighbour " + str(itr + 1) + " " + column + ","

        column_names += "\n"
        return column_names    

nearest_neighbors_content = getColumnNames()
for idx in range(0, len(indices)):
    source_stream_id = getTextForOutput(stream_tag_frequency_df[knn_columns_to_select].iloc[idx])
    print(source_stream_id.split(",")[0])
    nearest_neighbors_content += source_stream_id
    for neighbor_number, neighbor in enumerate(indices[idx]):
        distance = distances[idx, neighbor_number]
        
        if distance > min_distance_threshold and distance < max_distance_threshold:
            print("Distance: " + str(distance))
            nearest_neighbors_content += getTextForOutput(stream_tag_frequency_df[knn_columns_to_select].iloc[neighbor])
            
    nearest_neighbors_content += NEWLINE
    print(NEWLINE)
    
with open(nearest_neighbors_streams_output_file, "w") as fw:
    fw.writelines(nearest_neighbors_content)
            
```
