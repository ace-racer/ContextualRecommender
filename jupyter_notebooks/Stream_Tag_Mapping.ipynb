### Adds random tags to existing streams to augment existing data

```python
import random

## start of configurations

stream_tag_mapping_file_location = "../data/stream_tag_mapping_original.csv"
tags_file_location = "../data/tags.csv"
stream_types_to_exclude = ["nokia"]
stream_name_location = 2
stream_id_location = 1
MIN_TAGS = 2
MAX_TAGS = 6
## end of configurations


## start of constants

comma_str = ","
NEWLINE = "\n"

## end of constants

## start of global variables

new_content = ""
available_tags = []

## end of global variables

def does_stream_name_contain_specific_keywords(stream_name):
    """Check if the stream name contains specific keywords"""
    if stream_name:
        for item in stream_types_to_exclude:
            if item in stream_name:
                return True
    return False

## read all the available tags
with open(tags_file_location, "r") as fr:
    available_tags = [line.strip() for line in fr.readlines()]

with open(stream_tag_mapping_file_location, "r") as fr:
    for line_number, line in enumerate(fr.readlines()):
        print("Processing line: " + str(line_number))
        
        line = line.strip()
        line_contents = line.split(comma_str)
        
        num_tags_for_stream = random.randint(MIN_TAGS, MAX_TAGS)
        print("Number of tags for the stream: " + str(num_tags_for_stream))
        stream_name_lower = line_contents[stream_name_location].lower()
        
        # if the stream name does not contain any of the stream names to exclude
        if not does_stream_name_contain_specific_keywords(stream_name_lower):
            sampled_tags = random.sample(available_tags, num_tags_for_stream)
            for tag in sampled_tags:
                new_content += tag + comma_str + line_contents[stream_id_location] + comma_str + line_contents[stream_name_location] + NEWLINE

with open(stream_tag_mapping_file_location, "a") as fw:
    fw.writelines(new_content)
    
print("Content written to file...")


```
